# Tensor Math & Comparison Operators

```python
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])
```

```python
# Addition
z1 = torch.empty(3)
torch.add(x, y, out=z1)
```

```python
z2 = torch.add(x, y)
z2
```

```python
z = x+y
z
```

```python
# Subtraction
z3 = x-y
z3
```

```python
z4 = torch.true_divide(x, y)
z4
```

```python
# inplace operations
t = torch.zeros(3)
t.add_(x)            # t+= x
t
```

```python
# Exponentiation
z = x.pow(2)
z
```

```python
z = x ** 2
z
```

```python
# Simple Comparison
z = x > 2
z
```

```python
# Matrix Multiplication
x1 = torch.rand((2, 5))
x2 = torch.rand((5, 3))
x3 = torch.mm(x1, x2)
x3
```

```python
# Matrix exponentiation
m_exp = torch.rand(5, 5)
m_exp.matrix_power(3)
m_exp
```

```python
# Element wise Multiplication
z = x*y
z
```

```python
# Dot product
z = torch.dot(x, y)
z
```

```python
# Batch Matrix Multiplication
batch = 32
n = 10
m = 20
p = 30

tensor1 = torch.rand((batch, n, m))
tensor2 = torch.rand((batch, m, p))
out_bmm = torch.bmm(tensor1, tensor2)
out_bmm
```

## Broadcasting

```python
x1 = torch.rand((5, 5))
x2 = torch.rand((1, 5))

z = x1 - x2
z
```

# Other Tensor Operations

```python
x = torch.rand((1, 5))
x
```

```python
sum_x = torch.sum(x, dim=0)
sum_x
```

```python
val, i = torch.max(x, dim=0)
print(val, i)
```

```python
val, i = torch.min(x, dim=0)
print(val, i)
```

```python
abs = torch.abs(x)
abs
```

```python
z = torch.argmax(x, dim=0)
z = torch.argmin(x, dim=0)
```

```python
# mean requires float values
mean_x = torch.mean(x.float(), dim=0)
# check equality
z = torch.eq(x, y)
```

```python
sorted_x, i = torch.sort(x, dim=0, descending=False)
sorted_x
```

```python
# All elements in x less than zero will be set to zero
z = torch.clamp(x, min=0, max=10)
```

```python
x = torch.tensor([1,0,1,1,1], dtype=torch.bool)
# Atleast one value is true
z = torch.any(x)
z
```

```python
# No value is False
z = torch.all(x)
z
```
